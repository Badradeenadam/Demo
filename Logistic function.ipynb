```python
#logistic funtion 
#f(x;a,b,c,d)=a/1+exp(-c(x-d))+b
import numpy as np
from matplotlib import pyplot as plt
```

```python
def logistic_function(x,a=1, b=0, c=1, d=0):
    return a / (1+np.exp(-c*(x-d)))+b
```

```python
values=[-2, -1, 0, 1, 2]
x=np.linspace(-6,6)
```

```python
for a in values:
    plt.plot(x, logistic_function(x, a=a), label=f"a={a}")
   
```

```python
for a in values:
    plt.plot(x, logistic_function(x, b=a), label=f"b={a}")
   
```

```python
for a in values:
    plt.plot(x, logistic_function(x,c=a), label=f"c={a}")
   
```

```python
for a in values:
    plt.plot(x, logistic_function(x, d=a), label=f"d={a}")
   
```
